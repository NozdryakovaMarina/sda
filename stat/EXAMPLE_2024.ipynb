```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sci
import seaborn as sns

from scipy.stats import norm
```

```python
#исходные данные задачи 
##### PART 1
n=50
mu, sigma = 0, 1 
# mean and standard deviation
```

```python
s = np.random.normal(mu, sigma, n)
a_estimate=np.mean(s)
sigma_estimate=np.std(s,ddof=1)
(a_estimate,sigma_estimate)
print(s)
print(a_estimate)
```

```python
#### вывод ящика с усами
import seaborn as sb
W=sb.boxplot(s)
print(W)
```

```python
k_bins = 4
count, bin_edges = np.histogram(s,k_bins)
### count содержит абсолютные частоты для каждого интервала
print(count)
print(bin_edges)
sum(count)
```

```python
k_bins=4
x = np.linspace(-5, 5, 1000)
ax = norm.pdf(x, mu, sigma)
plt.plot(x, ax)
count, bins, ignored  = plt.hist(s, k_bins, density=True)
plt.show()
### гистограмма частот: сумма высот прямоугольников равна объему выборки
print(sum(count))
```

```python
y = norm.cdf(x)
plt.plot(x, y)
plt.show()
```
